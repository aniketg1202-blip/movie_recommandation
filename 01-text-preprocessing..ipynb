```python
def lower_replace(series):
    series = series.str.lower()
    series = series.str.replace(r"\[.*?\]", "", regex=True)
    series = series.str.replace(r"[^a-z0-9\s]", "", regex=True)
    return series
```

```python
df['sentence_f_cleaned'] = lower_replace(df['sentence'])
```

```python
import spacy as sp
nlp = sp.load('en_core_web_sm')
```

```python
text = 'when life gives you lemons make lemonade'
doc = nlp(text)
```

```python
def important_words(sentence):
    doc = nlp(sentence)
    words = []

    for token in doc:
        if not token.is_stop:
            words.append(token.lemma_)

    sentence = ' '.join(words)
    return sentence
```

```python
sentence = 'i was playing cricket there were many friends'
imp_words = important_words(sentence)
imp_words
```

```python
df['final_sentence'] = df['sentence_f_cleaned'].apply(important_words)
```

```python
df
```

```python
df = pd.DataFrame(data = data, columns = ['sentence'])
```

```python
def nlp_pipeline(series):
    series = lower_replace(series)
    series = series.apply(important_words)
    return series
```

```python
df['sentence_cleaned'] = nlp_pipeline(df['sentence'])
```
